```python
from diffusers import StableDiffusionImg2ImgPipeline
from PIL import Image
import torch

# Initialize pipeline (use float16 for reduced memory usage)
pipe = StableDiffusionImg2ImgPipeline.from_pretrained(
    "runwayml/stable-diffusion-v1-5",
    torch_dtype=torch.float16  # Use float16
).to("cuda")

# Upload your image (click folder icon on Colab's left sidebar)
init_image = Image.open("/content/1W9A3002.JPG").convert("RGB")

# Resize the image to a smaller size to reduce memory usage BEFORE the loop
# The model is typically trained on 512x512, so resizing to this size is a good starting point.
init_image = init_image.resize((512, 512))

# Generate 10 variants
for i in range(10):
    images = pipe(
        prompt="nuclear fuel bundle with cracks and rust, industrial lighting, high detail, 4K",
        image=init_image, # Use the resized image here
        strength=0.6,  # Balance between realism and variation
        guidance_scale=7.5
    ).images
    images[0].save(f"output_{i}.jpg")
    print(f"Saved output_{i}.jpg")

print("Done! Check the Colab files sidebar for your images.")
```

```python
from diffusers import StableDiffusionImg2ImgPipeline
from PIL import Image
import torch
import os # Import the os module

# Initialize pipeline (use float16 for reduced memory usage)
pipe = StableDiffusionImg2ImgPipeline.from_pretrained(
    "runwayml/stable-diffusion-v1-5",
    torch_dtype=torch.float16  # Use float16
).to("cuda")

# Upload your image (click folder icon on Colab's left sidebar)
init_image = Image.open("/content/1W9A3002.JPG").convert("RGB")

# Resize the image to a smaller size to reduce memory usage BEFORE the loop
# The model is typically trained on 512x512, so resizing to this size is a good starting point.
init_image = init_image.resize((512, 512))

# Generate 10 variants
for i in range(10):
    images = pipe(
        prompt="nuclear fuel bundle with cracks and rust, industrial lighting, high detail, 4K",
        image=init_image, # Use the resized image here
        strength=0.6,  # Balance between realism and variation
        guidance_scale=7.5
    ).images

    # --- Start of new code for handling existing files ---
    base_filename = f"output_{i}.jpg"
    save_path = base_filename
    counter = 1
    # Check if the file exists, if it does, append a counter to the filename
    while os.path.exists(save_path):
        filename, ext = os.path.splitext(base_filename)
        save_path = f"{filename}({counter}){ext}"
        counter += 1
    # --- End of new code ---

    images[0].save(save_path)
    print(f"Saved {save_path}")

print("Done! Check the Colab files sidebar for your images.")
```
